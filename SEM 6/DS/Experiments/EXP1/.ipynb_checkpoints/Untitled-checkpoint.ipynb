```python
# Looking for the missing data or cells in the dataset
df.isnull().sum()
```

```python
# Fixng the data type of the cells
df["type"] = df["type"].fillna("Unknown")
```

```python
# Handling all the nulls cells which are string, converting them to numeric values then calculating their median

numeric_cols = ["developer", "publisher"]

for col in numeric_cols:
    df[col] = pd.to_numeric(df[col], errors="coerce")

df["rating"] = df["rating"].fillna(df["rating"].mean())
df["numVotes"] = df["numVotes"].fillna(df["numVotes"].mean())
df["runtime"] = df["runtime"].fillna(df["runtime"].median())
df["startYear"] = df["startYear"].fillna(df["startYear"].median())
df["episodes"] = df["episodes"].fillna(0)
df["endYear"] = df["endYear"].fillna(0)
```
