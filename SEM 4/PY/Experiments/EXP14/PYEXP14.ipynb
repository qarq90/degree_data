What Is Python Matplotlib?

Matplotlib is a powerful and versatile open-source plotting library for Python, designed to help users visualize data in a variety of formats. Developed by John D. Hunter in 2003, it enables users to graphically represent data, facilitating easier analysis and understanding. If you want to convert your boring data into interactive plots and graphs, Matplotlib is the tool for you.

Bar Chart

```python
import matplotlib.pyplot as plt

categories = ['5"2', '5"4', '5"6', '5"8', '5"9']
values = [2, 15, 34, 10, 3]

plt.bar(categories, values, color='orange')
plt.title("Bar Graph Example")
plt.xlabel("Height in Inches")
plt.ylabel("Number of students")
plt.show()
```

Histogram

```python
import numpy as np

data = [48, 44, 40, 47, 61, 16, 54, 39, 73, 74, 93, 37, 57, 5, 59, 90, 45, 65, 2, 24, 8, 86, 10, 98, 73, 85, 40, 94, 36, 6, 16, 75, 56, 76, 53, 42, 100, 34, 26, 23]

plt.hist(data, bins=10, color='lightgreen', edgecolor='black')
plt.title("Histogram Example")
plt.xlabel("Value Ranges (Bins)")
plt.ylabel("Frequency")
plt.show()
```

Scatter Plot

```python
x = [8, 86, 10, 98, 73, 85, 40, 94, 36, 6, 16, 75, 56, 76, 53, 42, 100, 34, 26, 23, 16]
y = [48, 4, 89, 61, 52, 28, 44, 19, 23, 38, 73, 20, 31, 34, 1, 63, 72, 33, 32, 76, 74]

plt.scatter(x, y, color='red', marker='o')
plt.title("Scatter Plot Example")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.grid(True)
plt.show()
```

Pie Chart

```python
labels = ['Red Dead Redemption 2', 'Resident Evil Village', 'Call Of Duty Black Ops 3', 'Horizon Zero Dawn']
sizes = [70, 10.9, 43, 8.4]
colors = ['red', 'gold', 'orange', 'blue']

plt.pie(sizes, labels=labels, colors=colors, autopct='%1.1f%%', startangle=90)
plt.axis('equal')
plt.title("Pie Chart Example")
plt.show()
```
