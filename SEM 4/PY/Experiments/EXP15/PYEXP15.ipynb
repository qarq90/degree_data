```python
# Importing necessary libraries
import pandas as pd  # For data manipulation and analysis
import matplotlib.pyplot as plt  # For data visualization
import numpy as np  # For numerical operations

# Number of trials/samples we'll generate to demonstrate the Law of Large Numbers
num_trials = 100

# List to store the means of our samples
means = []

# Generate random samples and calculate their means
for i in range(1, num_trials+1):
    # Create a pandas Series of random integers between 1 and 10 (inclusive)
    # Size increases with each iteration (from 1 to num_trials)
    sample = pd.Series(np.random.randint(1, 11, size=i))  # Note: Changed to 11 since upper bound is exclusive

    # Calculate the mean of the current sample and store it
    means.append(sample.mean())

# Create a DataFrame to store our results
df = pd.DataFrame({
    "Sample_Size": range(1, num_trials+1),  # X-axis: Number of samples
    "Mean": means  # Y-axis: Calculated mean for each sample size
})

# Display the first few rows of our DataFrame
print(df.head())

# Create visualization
plt.figure(figsize=(12, 6))  # Set figure size

# Plot the sample means against sample size
plt.plot(df["Sample_Size"], df["Mean"], label='Sample Mean', color='red')

# Add a horizontal line at the expected value (mean of uniform distribution 1-10)
# The theoretical mean is (1+10)/2 = 5.5 (since 10 is inclusive)
plt.axhline(5.5, color='blue', linestyle='--', label='Expected Value (5.5)')

# Add labels and title
plt.xlabel("Number of Samples")
plt.ylabel("Average of Samples")
plt.title("Demonstration of Law of Large Numbers")

# Add legend and grid
plt.legend()
plt.grid(True)

# Show the plot
plt.show()
```
